### Data Preprocessing

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("housing.csv")
```

There are some values missing in the total_bedrooms column. To clean the dataset, I took any null values and filled them with the median.

```python
# Checking for missing values in the dataset
missing_values = df.isnull().sum()

# To clean the dataset I filled missing values using the median of 'total_bedrooms'
df['total_bedrooms'].fillna(df['total_bedrooms'].median(), inplace=True)

# Confirm missing values are handled
cleaned_missing_values = df.isnull().sum()

# Display a list of the updated columns displaying count of corrected column entries
missing_values, cleaned_missing_values
```

### Modeling

```python
# Explore statistics using describe function
statistics = df.describe()
print(statistics)

# Scatter plots between median house value and independent variables
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Scatter plot: Median House Value vs. Median Income
sns.scatterplot(x='median_income', y='median_house_value', data=df, ax=axes[0])
axes[0].set_title('Median House Value vs. Median Income')

# Scatter plot: Median House Value vs. Population
sns.scatterplot(x='population', y='median_house_value', data=df, ax=axes[1])
axes[1].set_title('Median House Value vs. Population')

# Scatter plot: Median House Value vs. Number of Households
sns.scatterplot(x='households', y='median_house_value', data=df, ax=axes[2])
axes[2].set_title('Median House Value vs. Number of Households')

plt.tight_layout()
plt.show()
```
